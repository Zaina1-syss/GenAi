```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langgraph.prebuilt import create_react_agent
from langchain_core.tools import tool
```

```python
# 2. Set your API Key (Paste your key here)
os.environ["GOOGLE_API_KEY"] = "Replace your API key"
```

```python
# 3. Setup the Brain (Cloud version)
# 'gemini-1.5-flash' is super fast and has a huge context window
llm = ChatGoogleGenerativeAI(model="gemini-3-flash-preview", temperature=0)
```

```python
# 4. Setup the "Hands" (Tool)
@tool
def magic_calculator(num: int) -> int:
    """Adds 100 to any number."""
    return num + 100

tools = [magic_calculator]
```

```python
# 5. Create and Run the Agent
agent = create_react_agent(llm, tools)

print("--- Starting Gemini Agent ---")
response = agent.invoke({"messages": [("user", "What is 50 + 100 using my magic calculator?")]})
```

```python
# Print only the final answer
print(f"Gemini: {response['messages'][-1].content}")
```
